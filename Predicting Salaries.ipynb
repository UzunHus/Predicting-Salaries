```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# veri setini pandasla alıp dataFrame nesnesi olan df'in içine aktaralım
df = pd.read_csv("polynomial.csv", sep =";")

```

```python
#veri setine bakalım
plt.scatter(df["deneyim"], df["maas"])
plt.xlabel("Deneyim (yıl)")
plt.ylabel("Maaş")
plt.savefig("1.png", dpi=300)
plt.show()
```

```python
# verilerin doğrusal yapıda dağılmadığı görülüyor
# eğer bu veri setine lineer regresyon uygulanırsa uygun olmayan bir tahmin elde edilir
reg = LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])

plt.xlabel("Deneyim (yıl)")
plt.ylabel("Maaş")

plt.scatter(df["deneyim"],df["maas"])

xekseni = df["deneyim"]
yekseni = reg.predict(df[["deneyim"]])
plt.plot(xekseni,yekseni, color="green", label = "linear regression")
plt.legend()
# bir adet polynomial regression nesnesi oluşturması için PolynomialFeatures fonksiyonunu çağıralım
# bu fonksiyonu çağırırken polinomun derecesini (N) belirtiyoruz:

polynomial_regression = PolynomialFeatures(degree =2 ) # degre = Nplt.show()
```

```python

# bir adet polynomial regression nesnesi oluşturması için PolynomialFeatures fonksiyonunu çağıralım
# bu fonksiyonu çağırırken polinomun derecesini (N) belirtiyoruz:
polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
# regression model nesnemiz olan reg nesenesini oluştuurp bunun fit metodunu çağğırarak x_polynomial ve y eksenlerini fit ediyor
# yani regresyon modelini mevcut gerçek verilerle eğitiyoruz :
reg = LinearRegression()
reg.fit(x_polynomial, df["maas"])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"], y_head, color="red", label = "polynomial regression")
plt.plot(xekseni, yekseni, color = "green", label = "Linear Regression")
plt.legend()

#veri setimizi de noktalar olarak scatter edelim ve polynomial regression uymuş bakalım
plt.scatter(df["deneyim"],df["maas"])

plt.show()
```
